### Data Cleaning 

```python
# rename SMS_received to SmsReceived
df.rename(columns={'SMS_received': 'SmsReceived'}, inplace=True)

# rename No-Show to NoShow
df.rename(columns={'No-show': 'NoShow'}, inplace=True)

# confirm changes
df.head(1)
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

df['AppointmentDay'] = np.array(df['AppointmentDay'].values, dtype='datetime64[D]')  
df['ScheduledDay'] = np.array(df['ScheduledDay'].values, dtype='datetime64[D]') 

df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

#confirm changes for 'ScheduledDay' and 'AppointmentDay' column
df.head(1)
```

```python
# confirm changes for data types
df.dtypes
```

```python
#Cheking the difference between appointment Day and Scheduled Day
df['RestDaysToApp'] = (df['AppointmentDay'] - df['ScheduledDay']) / np.timedelta64(1, 'D')

df['RestDaysToApp'].dtypes
```

```python
# Converting the type of RestDaysToApp from ' float64' to 'int64'.
df['RestDaysToApp'] = df['RestDaysToApp'].astype("int64")
df['RestDaysToApp'].dtypes
```

```python
# Check uniques of 'RestDaysToApp' column to catch the unmeaningful values.
df.RestDaysToApp.unique()
```

I see the unmeaningful results like '-6' and '-1'. The differences can not be the number that is under the '0'.So i have to drop them.

```python
#Dropping the negative results
df.drop(df[df['RestDaysToApp']<0].index,inplace=True)
```

```python
# Confirm changes
df.info()
```

```python
#Check outliers of Age
df_under_zero = df.query('Age < 0')
df_under_zero.head()
```

```python
#dropping outliers of Age values that is less than "0".
df.drop(df[df['Age']<0].index,inplace=True)


#confirm changes
df_under_zero = df.query('Age < 0')
df_under_zero.head()
```

## Explarotary Data Analysis

### Question 1( What is the relationship between age and no-show?)

```python
# Check the average of Ages regarding the No Showed results.
df_age=df.groupby('NoShow')['Age'].mean()
df_age
```

The average age of those who did not attend/miss their appointment is younger than the average age of those who go.

```python
#Plot the histogram to understand the distributions of Ages the No-Show situation 
df.query('NoShow=="Yes"' ).Age.hist(alpha=0.5, bins=30,label="missed/didn't come")
df.query('NoShow=="No"' ).Age.hist(alpha=0.5, bins=30,label="camed")
plt.legend();
plt.title("Age histogram of the patient that camed or missed/didn't come to the appointment")
plt.xlabel("Ranges of Age ")
plt.ylabel("Count of results")
plt.show();
```

The number of coming to appointments is higher in individuals over the age of 40.
We can say that the number of patients who come to appointments for 0-year-old babies is overwhelmingly superior to those who do not.
After the age of 60, the number of missed appointments has decreased.

```python
#I checked the counts of the ages belons patient.
df.Age.value_counts().plot(kind='bar',figsize=(16,16))
plt.show()
```

The ages with the highest number of patients are 0 and 1 year old babies.

### Research Question 2  ( What is the effect of reporting the appointment by sms in different age groups on the NoShow value?)

```python
#Plot an histogram to understand the differences between camed and didn't come of patient after sms notification.
df.query('NoShow=="Yes" &  SmsReceived==1').Age.hist(alpha=0.5, bins=30,label="missed/didn't come")
df.query('NoShow=="No"  & SmsReceived==1').Age.hist(alpha=0.5, bins=30,label="camed")
plt.legend();
plt.title("Age histogram of the patient that camed or missed/didn't come to the appointment after they received sms notification")
plt.xlabel("Ranges of Age ")
plt.ylabel("Count of results")
plt.show();
```

It can be said that receiving an SMS notification is effective in the number of people over the age of age coming to an appointment.However, a more detailed investigation is required.

```python
# I adapted the code that i found from a website to make groups from Ages and create a new column in No-Show Data.
#Source:https://stackoverflow.com/questions/62768980/pandas-split-ages-by-group
def AgeGrouper(df):
    df["AgeGroups"] = pd.cut(x=df['Age'], bins=[0,2,18,40,60,90,120], labels=["Baby","Children","Young","MiddleAged","Old","ExtraOld"])
    return df

# Call function to group ages and create a new column in No-Show Data
df=AgeGrouper(df)
```

```python
#I filtered the data of the people who did not come to the appointment using query and worked with this data frame.
df_No_Showed=df.query('NoShow=="Yes"' )

# I grouped the data I filtered according to age groups and looked at the percentage of receiving SMS notifications of those in these groups.
# I plotted these percentages to see them better.

df_No_Showed.groupby('AgeGroups')['SmsReceived'].mean().plot()
plt.title("Percentage chart for the patients who recieved Sms and didn't show upped according to the age groups ")
plt.xlabel("Age Groups")
plt.ylabel("Percentages")
plt.show()
```

We see that the highest rate of not receiving an SMS notification is in the infant and young age groups.

```python

# I calculated the percentage of receiving sms notifications in patients from different age groups who came to their appointments.
# I have plotted these percentages to see them better.

df_No_Showed=df.query('NoShow=="No"' )
df_No_Showed.groupby('AgeGroups')['SmsReceived'].mean().plot()
plt.title("Percentage chart for the patients who recieved Sms and show upped according to the age groups ")
plt.xlabel("Age Groups")
plt.ylabel("Percentages")
plt.show()
```


The proportion of people who receive SMS notifications is the highest in the middle age group.

### QUESTION 3(What is the effect of hypertension on the absence of appointments from patients who have received SMS notifications according to age groups?)

```python
# Plot the status of hypertension according to age groups of patients who do not attend/miss an appointment.
df_No_Showed=df.query('NoShow=="Yes"' )
df_No_Showed.groupby('AgeGroups')['Hipertension'].mean().plot()
plt.title("Percentage chart for the patients who has hipertension and missed/didn't come to an appointment according to the age groups ")
plt.xlabel("Age Groups")
plt.ylabel("Percentages of having hipertension")
plt.show()
```

The percentage of old and extra old patients who do not come to an appointment to have hypertension disease is over 60 percent.

```python
df_No_Show=df.query('NoShow=="No"' )
df_No_Show.groupby('AgeGroups')['Hipertension'].mean().plot()
plt.title("Percentage chart for the patients who has hipertension and camed to an appointment according to the age groups ")
plt.xlabel("Age Groups")
plt.ylabel("Percentages of having hipertension")
plt.show()
```

The percentage of old and extra old patients who come to an appointment to have hypertension disease is over 55 percent.
